```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

Faça a leitura dos dados do arquivo CSV:


Altere o código abaixo de acordo com seu diretório.

```python
df = pd.read_csv("MODULO7_PROJETOFINAL_BASE_SUPERMERCADO.csv")

df.head(10)
```

Os campos do nosso dataframe são:

**Title:** Nome do produto.


**Marca:** A marca do produto.


**Preco_Normal:** O preço em que o produto costuma ser vendido quando não há desconto.


**Preco_Desconto:** O preço vendido após o desconto ser aplicado.


**Preco_Anterior:** Preço em que era comercializado o produto antes do desconto aplicado.


**Desconto:** Total de desconto aplicado.






As colunas que aparecem com valores 0 são para os produtos onde não tivemos descontos aplicados.


As categorias estão em espanhol!

# 1 - Traga a média e a mediana dos preços - coluna Preco_Normal - por categoria de produto.
# Identifique as categorias que parecem ter um valor de média abaixo ou acima da mediana.

```python
#Seu código aqui para a média
media_total = df['Preco_Normal'].mean().round(2)
print(media_total)
df.groupby('Categoria')['Preco_Normal'].mean().reset_index().round(2).sort_values(by='Preco_Normal', ascending=False)
```

```python
#Seu código aqui para a mediana
media_total = df['Preco_Normal'].median()
print(media_total)
df.groupby('Categoria')['Preco_Normal'].median().reset_index().sort_values(by='Preco_Normal', ascending=False)
```

Digite aqui as categorias:
Todas as categorias apresentam um deslocamento considerável em relação a média, contudo a mais discrepante das categorias é a dos lácteos. Sua média é quase 150% maior que sua mediana. Já é possível notar a presença de outliers nesta categoria.

# 2 - Traga o desvio padrão por categoria de produto.
# Qual o comportamento da média e mediana nas categorias com maior desvio?

```python
#Seu código aqui
desvio_padrao_por_categoria = df.groupby('Categoria')['Preco_Normal'].std().round(2).reset_index().sort_values(by='Preco_Normal', ascending=False)
desvio_padrao_por_categoria
```

Digite nesse campo o comportamento que você identificou.<br>
Os valores altos no desvio padrão confirmam a presença de outliers em algumas categorias. Destaque para os lacteos e instantaneo-y-sopa que estão com desvio bem acima das médias e medianas.

# 3 - Plot um boxplot da distribuição do Preco_Normal para a categoria que você identificou que tem o maior desvio padrão. Como é a distribuição desses dados segundo o boxplot? Você identifica muitos outliers?

Dica: Para trazer apenas os dados da categoria que você deseja você pode usar o df.loc[df['Categoria'] == 'CATEGORIA ESCOLHIDA'

```python
#Seu código aqui
lacteos = df.loc[df['Categoria'] == 'lacteos'] # Criação da variável lacteos para trazer somente os dados da categoria lacteos

fig = px.box(lacteos, y='Preco_Normal', title='Boxplot da Categoria lacteos')
fig.update_layout(xaxis_title='Lácteos', yaxis_title='Preco Normal')

fig.show()
```

# 4 - Plote um gráfico de barras onde temos a média de descontos por categoria.

```python
# Seu código aqui
media_desconto_por_categoria = df.groupby('Categoria')['Desconto'].mean().round(2).reset_index().sort_values(by='Desconto', ascending=False) # Cálculo da médiade desconto por categoria

fig = px.bar(media_desconto_por_categoria, x='Categoria', y='Desconto', title='Média de Desconto por Categoria')
fig.update_layout(xaxis_title='Categoria', yaxis_title='Média de Desconto')
```

# 5 - Plote um gráfico de mapa interativo agrupando os dados por categoria, marca e trazendo a média de desconto.

```python
# Seu código aqui
media_desconto = df.groupby(['Categoria', 'Marca'], as_index=False)['Desconto'].mean() # Calculo da média de desconto por categoria e marca

fig = px.treemap(media_desconto, path=['Categoria', 'Marca'], values='Desconto', title='Média de Desconto por Categoria e Marca')

fig.show()
```
